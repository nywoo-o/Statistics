```python
googlehdiff = google['High'].diff().iloc[1:]
googlevdiff =  google['Volume'].diff().iloc[1:]

plot_acf(googlehdiff, lags=25, title="google High diff acf")
plt.show()
```

```python
plot_pacf(googlevdiff, lags=25, title="google volume diff pacf")
plt.show()
```

google high diff ARMA model

```python
googleHighARMA = ARMA(googlehdiff, order=(1, 0))
res = googleHighARMA.fit()
ax = plt.subplot()
ax = googlehdiff.loc['2016':].plot(ax=ax)
res.plot_predict(start=2700, ax=ax, plot_insample=False)
plt.show()

print(res.summary())

rmse = math.sqrt((res.resid**2).mean())
print("Root mean squared error is {}.".format(rmse))

rmse = math.sqrt((res.resid**2).mean())/googlehdiff.var()
print("N Root mean squared error is {}.".format(rmse))

print(googlehdiff.var())
plot_acf(res.resid, lags=25)
plt.show()


print(normaltest(res.resid))
```
